```python
import pandas as pd
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import A4

planilha = pd.read_excel('/content/FuncionáriosAtuais.xlsx')
tabela = planilha.head(5)

# Convert Excel serial dates to datetime objects
tabela['Data de Contratacao'] = pd.to_datetime(tabela['Data de Contratacao'], unit='D', origin='1899-12-30')
tabela['Data de Demissao'] = pd.to_datetime(tabela['Data de Demissao'], unit='D', origin='1899-12-30', errors='coerce')
tabela['Data de Nascimento'] = pd.to_datetime(tabela['Data de Nascimento'], unit='D', origin='1899-12-30')

tabela
```

```python
tabela['CPF'] = tabela['CPF'].astype(str)
tabela['RG'] = tabela['RG'].astype(str)
tabela['ID RH'] =  tabela['ID RH'].astype(str)
tabela['ID de Pessoal'] = tabela['ID de Pessoal'].astype(str)
tabela['CEP'] = tabela['CEP'].astype(str)
tabela['Ramal'] = tabela['Ramal'].astype(str)
```

```python

for linha in tabela.index:

# criar relatório

  nome = tabela.loc[linha, "Nome Completo"]
  id_rh = tabela.loc[linha, "ID RH"]
  rg = tabela.loc[linha, "RG"]
  cpf = tabela.loc[linha, "CPF"]
  ramal = tabela.loc[linha, "Ramal"]
  estado_civil = tabela.loc[linha, "Estado Civil"]
  login = tabela.loc[linha, "Login"]
  data_nasc = tabela.loc[linha, "Data de Nascimento"]
  cep = tabela.loc[linha, "CEP"]
  data_contratacao = tabela.loc[linha, "Data de Contratacao"]
  dias_uteis = tabela.loc[linha, "Dias Uteis Trabalhados Ano Orcamentario"]
  salario_base = tabela.loc[linha, "Salario Base"]
  impostos = tabela.loc[linha, "Impostos"]
  beneficios = tabela.loc[linha, "Beneficios"]
  vt = tabela.loc[linha, "VT"]
  vr = tabela.loc[linha, "VR"]
  cargo = tabela.loc[linha, "Cargo"]
  bandeira = tabela.loc[linha, "Bandeira"]
  codigo = tabela.loc[linha, "Codigos"]
  q_acessos = tabela.loc[linha, "Quantidade de Acessos"]
  ferias_acumuladas = tabela.loc[linha, "Ferias Acumuladas"]
  ferias_remuneradas = tabela.loc[linha, "Ferias Remuneradas"]
  hora_extra = tabela.loc[linha, "Horas Extras"]
  valores_adicionais = tabela.loc[linha, "Valores Adicionais"]
  id_de_pessoal = tabela.loc[linha, "ID de Pessoal"]
  id_de_area = tabela.loc[linha, "ID da area"]

  documento_pdf = canvas.Canvas(f"Relatório de Funcionários Atuais: {nome}.pdf", A4)

  # título
  # escolher fonte
  documento_pdf.setFont("Helvetica-Bold", 12)
  documento_pdf.drawString(100, 750, f"Relatório de Funcionários Atuais: {nome}")

  # escrever o texto

  documento_pdf.setFont("Helvetica-Bold", 9)
  d = 20
  documento_pdf.drawString(100, 700, f"Nome: {nome}")
  documento_pdf.drawString(100, 700 - 1*d, f"ID RH: {id_rh}")
  documento_pdf.drawString(100, 700 - 2*d, f"RG: {rg}")
  documento_pdf.drawString(100, 700 - 3*d, f"CPF: {cpf}")
  documento_pdf.drawString(100, 700 - 4*d, f"Ramal: {ramal}")
  documento_pdf.drawString(100, 700 - 5*d, f"Estado Civil: {estado_civil}")
  documento_pdf.drawString(100, 700 - 6*d, f"Login: {login}")
  documento_pdf.drawString(100, 700 - 7*d, f"Data de Nascimento: {data_nasc}")
  documento_pdf.drawString(100, 700 - 8*d, f"CEP: {cep}")
  documento_pdf.drawString(100, 700 - 9*d, f"Data de Contratacao: {data_contratacao}")
  documento_pdf.drawString(100, 700 - 10*d, f"Dias Úteis: {dias_uteis}")
  documento_pdf.drawString(100, 700 - 11*d, f"Salário Base: {salario_base}")
  documento_pdf.drawString(100, 700 - 12*d, f"Impostos: {impostos}")
  documento_pdf.drawString(100, 700 - 13*d, f"Benefícios: {beneficios}")
  documento_pdf.drawString(100, 700 - 14*d, f"VT: {vt}")
  documento_pdf.drawString(100, 700 - 15*d, f"VR: {vr}")
  documento_pdf.drawString(100, 700 - 16*d, f"Cargo: {cargo}")
  documento_pdf.drawString(100, 700 - 17*d, f"Bandeira: {bandeira}")
  documento_pdf.drawString(100, 700 - 18*d, f"Código: {codigo}")
  documento_pdf.drawString(100, 700 - 19*d, f"Quantidade de Acessos: {q_acessos}")
  documento_pdf.drawString(100, 700 - 20*d, f"Férias Acumuladas: {ferias_acumuladas}")
  documento_pdf.drawString(100, 700 - 21*d, f"Férias Remuneradas: {ferias_remuneradas}")
  documento_pdf.drawString(100, 700 - 22*d, f"Hora Extra: {hora_extra}")
  documento_pdf.drawString(100, 700 - 23*d, f"Valores Adicionais: {valores_adicionais}")
  documento_pdf.drawString(100, 700 - 24*d, f"ID de Pessoal: {id_de_pessoal}")
  documento_pdf.drawString(100, 700 - 25*d, f"ID de Área: {id_de_area}")
  documento_pdf.drawString(100, 700 - 26*d, f"Esse Relatório foi gerado automaticamente com Python")



  # salvar os relatórios
  documento_pdf.save()
```
